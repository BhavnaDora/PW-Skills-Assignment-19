#### (1)-
##### GET and POST are the methods of HTTP(Hyper Text Transfer Protocol) which is used to create web APIs. Both of them are used to request data. 
##### GET is used to request data and this data is passed through URL from client side to server side. It is the default method of requesting data. GET requests can be cached and they remain in browser history, so they should never be used while dealing with sensitive data. Ex) Google search
##### POST method is used to send data from client side to server side to create/update a resource. This data is passed through a form. POST requests are never cached and they do not remain in browser history, so they should be used while dealing with sensitive information. Ex) Gmail

#### (2)-
##### Request is an object in Flask that contains all the data sent from client to server. This data is recovered using the GET/POST methods. It has many importnat attributes- Form, args, method, files, Cookies.

#### (3)-
##### Redirect() is a function in Flask. When its called, it returns a response object and redirects the user to another target location with the specified status code. It takes 3 parameters- location is the url where the response should be redirected, status code sent to browser's header defaults to 302, response parameter is used to instantiate response.

```python
from flask import Flask, request, jsonify
app1= Flask(__name__)

@app1.route("/postman", methods= ['POST'])
def eligibile():
    if request.method== 'POST':
        age= request.json['age']
        nation= request.json['nationality']
        try:
            if age>=18 and age<=100:
                if nation== 'India':
                    res= "you are citizen of nation {} and your age {} is eligible to vote".format(nation,age)
        except Exception as e:
            res= "you are not eligible to vote"
        return jsonify(res)

if __name__=="__main__":
    app1.run(host="0.0.0.0")
```
